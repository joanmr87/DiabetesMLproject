```python
#Importar los datos y las librerias necesarias.
#https://www.kaggle.com/datasets/alexteboul/diabetes-health-indicators-dataset?resource=download
import pandas as pd
import numpy as np
diabetis = pd.read_csv("diabetes.csv")
diabetis
```

```python
#cargar el modelo
from sklearn.ensemble import RandomForestClassifier
modelo = RandomForestClassifier()
```

```python
#separar en dos la tabla, por un lado los datos y por otro los resultados.
datos = diabetis.drop("Diabetes_binary", axis=1)
resultados = diabetis["Diabetes_binary"]
```

```python
datos
```

```python
resultados
```

```python
#obtener juegos de datos para entrenar el modelo.
from sklearn.model_selection import train_test_split
#nos devuelve 4 juegos de datos, 2 de entreno y dos de test
datos_entreno, datos_test, resultados_entreno, resultados_test = train_test_split(datos, resultados, test_size=0.2)
```

```python
datos_entreno
```

```python
datos_test
```

```python
#entrenamos el modelo
modelo.fit(datos_entreno, resultados_entreno)
```

```python
#usando el modelo, crearemos algunas predicciones con los datos de test
predicciones = modelo.predict(datos_test)
predicciones
```

```python
#obtener la fiabilidad del modelo, comparando los datos de test conocidos
modelo.score(datos_test, resultados_test)
#Exactitud (accuracy). Compara todos los datos
```

```python
#Obtener otra metrica. Recall / Exaustividad. Tasa de aciertos en positivos.
from sklearn.metrics import recall_score
recall = recall_score(resultados_test, predicciones)
recall
```

```python
#Diferentes variables que se pueden modificar para mejorar el modelo.
modelo.get_params()
```

```python
#Creamos un bucle que evalue el modelo
for i in range (50, 210, 10): 
    print(f"n_estimator: {i}")
    modelo = RandomForestClassifier(n_estimators=i)
    modelo.fit(datos_entreno, resultados_entreno)
    print(f"Exactitud / Accuracy {modelo.score(datos_test, resultados_test)}")
    predicciones = modelo.predict(datos_test)
    print(f"Exaustividad / Recall {recall_score(resultados_test, predicciones)}")
    print("--------")
```
